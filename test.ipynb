```python
import pint
import scipy
from matplotlib import pyplot as plt
import numpy as np
from open_atmos_jupyter_utils import show_plot
```

```python
si = pint.UnitRegistry()
```

```python
R_str = scipy.constants.R * si.J / (si.mol * si.K)
```

```python
M = 28.966 * si.g / si.mol
R = R_str / M
p = 101325 * si.Pa
T = 293.15 * si.K
```

```python
def rho(p, T):
    return p / (R * T)
```

```python
rho_pow = rho(p, T).to_base_units()
print(rho_pow)
```

```python
print(f"{rho_pow.to(si.lb / si.inch**3)=:.3g}")
```

```python
si.setup_matplotlib()
```

```python
plt.plot(T:=np.linspace(200, 300) * si.K, rho(p, T).to_base_units(), label="gęstość")
plt.legend()
# plt.show()
show_plot()
```
